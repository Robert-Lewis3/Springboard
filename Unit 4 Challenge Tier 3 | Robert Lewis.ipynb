**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def Housing_Price_Ratio(d):
    
    # Looks at input arg, finds corresponding average prices for 1999 & 2019
    yr1999 = float(d['Avg_Price'][d['Year'] == 1999])  
    yr2019 = float(d['Avg_Price'][d['Year'] == 2019])  
    # Convert yr1998 & yr2008 to a ratio
    ratio  = yr1999 / yr2019
    # Return the ratio for given input
    return ratio
```

```python
# Test Function using the data from 'Ealing'

Housing_Price_Ratio(dfgroup[dfgroup['London_Boroughs']=='Ealing'])
```

```python
# To make a final df of all the ratios found across all London
# boroughs, start with initializing an empty dictionary
# which will be populated using a 'for' loop

final = {}
```

```python
# Create a for loop to find 1998/2008 property value ratios 
# across all of the London borroughs

# Use x as an iterator over the list of unique values in the 'London_Boroughs' column
for x in dfgroup['London_Boroughs'].unique():
    # Create variable that is the subset parameter to the Housing_Price_Ratio function  
    borough = dfgroup[dfgroup['London_Boroughs'] == x]
    # Return the Housing_Price_Ratio for the given borough and assign to final dictionary
    final[x] = Housing_Price_Ratio(borough)
    
print(final)
```

```python
df_ratios = pd.DataFrame(final.items())
```

```python
df_ratios.head()
```

```python
df_ratios = df_ratios.rename(columns= {0: 'Borough', 1: 'Ratio'})
df_ratios.tail()
```

```python
df_ratios['Percent_Increase'] = round((df_ratios['Ratio'] * 100), 2)
```

```python
# Sort boroughs by descending ratios and print the Top 10

top20 = df_ratios.sort_values(by='Ratio',ascending=False).head(20)
print(top20)
```

```python
# To look at the spread, the bottom 10 were also tabulated

bottom10 = df_ratios.sort_values(by='Ratio',ascending=False).tail(10)
print(bottom10)
```

```python
TopPlotBar = top20[['Borough','Ratio']].plot(kind='bar')

TopPlotBar.set_xticklabels(top20.Borough)
```

In April of 1965, the city of London decided to reorganize the city's various districts into 32 boroughs. The boroughs can be subdivided into two main groups: Inner London & Outer London. Boroughs in inner London tend to be smaller and more densely populated, whereas outer boroughs tend to be larger and more sparse in population (Source: Wiki London boroughs page). 

The dataset for this project focused on housing prices, in London, for the years 1995 - 2020. From this dataset, we looked to find the boroughs whose housing prices increased the most over the last two decades.

Interestingly, with the exception of 'Hammersmith & Fulham', average price increases were higher for boroughs in outer London. Some contributing factors could be: 1) Young families wanting to spread out 2) Less crime in outer boroughs 3) Middle age/older citizens leaving denser inner London

The exception of Hammersmith & Fulham isn't totally surprising, given that it is amoung the most expensive markets for residential real estate in London. H&F is home to many international corporations, 3 professional football teams and contains a major roadway (A4) that connects London to Bristol via Heathrow Int'l airport.
